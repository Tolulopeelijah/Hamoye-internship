```python
def replace_space(i):
    if len(i.strip()) == 0:
        return 0
    else: return float(i)


train['TotalCharges'] = train['TotalCharges'].apply(replace_space)
```

```python
train.drop(['customerID'], axis = 1, inplace = True)
```

```python
categorical, numerical = [], []
for i in train.columns:
    if train[i].dtype == 'O':
        categorical.append(i)
    else:
        numerical.append(i)
categorical.pop(categorical.index('Churn'))
categorical.append(numerical.pop(numerical.index('SeniorCitizen')))
print(categorical)
print(numerical)
# categorical = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
# 'MultipleLines', 'InternetService',
# 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 
# 'StreamingMovies','Contract', 'PaperlessBilling', 'PaymentMethod']
# numerical = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
for i in categorical:
    print(f'{i} - {train[i].nunique()} unique values')
```

```python
train[numerical].describe()
```

```python
ss = StandardScaler()
train[['tenure', 'MonthlyCharges', 'TotalCharges']] = ss.fit_transform(train[['tenure', 'MonthlyCharges', 'TotalCharges']])
```

```python
train[numerical].describe()
```

```python
for i, j in enumerate(categorical):
    train[[j]].groupby(train[j]).count().plot(kind = 'bar')
```

```python
en = OneHotEncoder(sparse = False)
le = LabelEncoder()
train.Churn = le.fit_transform(train['Churn'])
encoded = en.fit_transform(train[categorical])
```

```python
encode_df = pd.DataFrame(encoded, columns = en.get_feature_names_out())
```

```python
encode_df.shape
```

```python
train_en = pd.concat([train, encode_df], axis = 1)
# pd.merge(train, objs, how='outer', on=index)
train_en.drop(categorical, axis = 1, inplace = True)
```

```python
y = train_en.pop('Churn')
X = train_en
```

```python
X_train, X_test,  y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
train_en
```

```python
model1 = RandomForestClassifier(random_state=1)
model2 = ExtraTreesClassifier(random_state=1)
model3 = xgboost.XGBClassifier(random_state=1)
model4 = lightgbm.LGBMClassifier(random_state=1)


model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)


prediction1 = model1.predict(X_test)
prediction2 = model2.predict(X_test)
prediction3 = model3.predict(X_test)
prediction4 = model4.predict(X_test)
```

```python
for i,j in zip([prediction1, prediction2, prediction3, prediction4], ['RandomForestClassifier', 'ExtraTreesClassifier', 'xgboost', 'lightgbm']):
    print(j)
    print(f'recall score: {recall_score(i, y_test)}')
    print(f'precision score: {precision_score(i, y_test)}')
    print(f'f1 score: {f1_score(i, y_test)}')
    print(f'accuracy score: {accuracy_score(i, y_test)}')
    print('\n')
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [50, 100, 300, 500, 1000]

min_samples_split = [2, 3, 5, 7, 9]

min_samples_leaf = [1, 2, 4, 6, 8]

max_features = ['auto', 'sqrt', 'log2', None] 

hyperparameter_grid = {'n_estimators': n_estimators,

                       'min_samples_leaf': min_samples_leaf,

                       'min_samples_split': min_samples_split,

                       'max_features': max_features}
```

```python
rf = RandomForestClassifier()
random_search = RandomizedSearchCV(rf, param_distributions=hyperparameter_grid, n_iter=10, cv=5,
                                   n_jobs=-1, random_state=1, scoring = 'accuracy',
                                  verbose = 1)
random_search.fit(X_train, y_train)
```
